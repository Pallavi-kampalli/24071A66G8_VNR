```python
a = input(" Enter a value: ")
a = int(a)
count = 0
for i in range(1, a + 1, 1):
  if (a % i == 0):
    count+=1
if (count == 2):
   print("The given number is prime.")
else:
  print("The given number is not prime.")
```

```python
a = input(" Enter a value: ")
a = int(a)
def fib(num):
  if (num == 0 or num == 1):
   return num
  else:
   return fib(num-1) + fib(num-2)

print(f"The fibonacci series for {a} is as follows")
for i in range(0 ,a+1 ,1):
  print(fib(i))
```

```python
print(" Enter two values a,b")
a = input()
b = input()
a = int(a)
b = int(b)
min = a if (a < b) else b
max = a if (a > b) else b
for i in range(1, min+1, 1):
   if( (a % i == 0 ) and (b % i == 0) ):
    GCD = i

for j in range(max, (a*b)+1, 1):
  if( (j % a == 0) and (j % b == 0)):
    LCM = j

print(f"The GCD of {a} and {b} is {GCD}")
print(f"The LCM of {a} and {b} is {LCM}")
```

```python
print(" Enter two values a,b")
a = input()
b = input()
a = int(a)
b = int(b)
print("Enter the operator")
c = input()
c = str(c)
def calculator(c):
    if c == '+':
        return a + b
    elif c == '-':
        return a - b
    elif c == '*':
        return a * b
    elif c == '/':
        return a / b
    elif c == '%':
        return a % b
    else:
        return "Invalid operator"

print(calculator(c))
```
